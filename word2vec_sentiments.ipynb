```python
import os
import pandas as pd
os.chdir("/Users/snerur/ericsson/cohort1_data")
with open("labels.txt") as infile:
    labels = infile.readlines()
with open("reviews.txt") as infile:
    reviews = infile.readlines()
```

```python
len(reviews)
```

```python
labels = [word.strip() for word in labels] #strip out newline characters
labels[:5]
```

```python
reviews = [word.strip() for word in reviews]
#let us put them in a Pandas Dataframe to see what they look like
df = pd.DataFrame()
df["Review"] = reviews
df["Label"] = labels
df.head()
```

```python
import gensim, logging
import os
import glob
from gensim.summarization import summarize
from gensim.summarization import keywords


class MySentences(object):
    def __init__(self, corpus):
        self.corpus = corpus
 
    def __iter__(self):
        for line in self.corpus:
            yield line.split()
                
                
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)

model = gensim.models.Word2Vec(MySentences(reviews), min_count=10, iter = 500)
```

```python
#let us check similarity of some word, say, "brilliant"
model.wv.most_similar("brilliant")
```
